# 次元削減
iris / wine / digits の各データセットの諸元は以下のとおり：
- iris dataset : 4次元 / 3クラス
- wine dataset : 13次元 / 3クラス
- digits dataset : 64次元 / 10クラス

これらを２次元へ次元削減し，可視化してみよう。

目的：
- 次元削減の基本的な手法として PCA の使い方をマスターする。
- 近年では Isomap や t-SNE などの非線形手法が用いられることを知る。

---
## 主成分分析 (PCA) を用いた次元削減
主成分分析(PCA; Principal Component Analysis)とは，各主成分の分散が最大になるよう，主成分を定める手法である。

主成分分析の解釈について，以下のサイト「主成分分析の考え方」が分かりやすい。（ただし，コードは R言語で書かれている)  
https://logics-of-blue.com/principal-components-analysis/


(参考) sklearn のおける PCA のマニュアル:  
https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
'''
PCA を用いた次元削減
'''
# データセットの読み込み
from sklearn.datasets import load_iris
dataset = load_iris()
 
#データの整理
target_names = dataset.target_names#ターゲット(花の種類)の名前リスト
targets = dataset.target#ターゲットに与えられた番号
feature_names = dataset.feature_names#特徴名リスト
features = dataset.data#各特徴のデータ

# PCA(主成分分析)を用いた次元削減
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transformed = pca.fit_transform(features)#featuresを2次元に変換

# 可視化
import numpy as np
import matplotlib.pyplot as plt
for label in np.unique(targets):
   plt.scatter(transformed[targets==label, 0], transformed[targets==label, 1], label=target_names[label])

plt.legend()
plt.show()
```

## 次元削減に PCA を用いるのは古いか？
高次元データの次元削減法を以下にまとめる。
- 1900〜1960年: 主成分分析 (PCA) ・・・固有値問題を解く
- 1990〜2000年: Isomap ・・・非線形
- 2000年〜現在: t-SNE (T-distributed Stochastic Neighbor Embedding), UMAP

(参考) Isomap のマニュアル:  
https://scikit-learn.org/stable/modules/generated/sklearn.manifold.Isomap.html

(参考) t-SNE のマニュアル:  
https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html

以下では，64次元 / 10クラスの digits dataset を対象に次元削減を試みる。

```python
'''
Isomap を用いた次元削減
'''
# データセットの読み込み
from sklearn.datasets import load_digits
dataset = load_digits()
 
#データの整理
target_names = dataset.target_names#ターゲット(花の種類)の名前リスト
targets = dataset.target#ターゲットに与えられた番号
feature_names = dataset.feature_names#特徴名リスト
features = dataset.data#各特徴のデータ

# t-SNE を用いた次元削減
from sklearn.manifold import Isomap
isomap = Isomap(n_components=2)
transformed = isomap.fit_transform(features)#featuresを2次元に変換

# 可視化
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8)) # 図を大きめにする
for label in np.unique(targets):
   plt.scatter(transformed[targets==label, 0], transformed[targets==label, 1], label=target_names[label])

plt.legend()
plt.show()
```

```python
'''
t-SNE を用いた次元削減
'''
# データセットの読み込み
from sklearn.datasets import load_digits
dataset = load_digits()
 
#データの整理
target_names = dataset.target_names#ターゲット(花の種類)の名前リスト
targets = dataset.target#ターゲットに与えられた番号
feature_names = dataset.feature_names#特徴名リスト
features = dataset.data#各特徴のデータ

# t-SNE を用いた次元削減
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
transformed = tsne.fit_transform(features)#featuresを2次元に変換

# 可視化
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8)) # 図を大きめにする
for label in np.unique(targets):
   plt.scatter(transformed[targets==label, 0], transformed[targets==label, 1], label=target_names[label])

plt.legend()
plt.show()
```
